## Title -To study Sets in Python
Sets are used to store multiple items in single variable.

Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List,Tuple,and Dictionary,all with different qualities and usage.
A set is a collection which is unordered,unchangeable*,and unindexed.

```python
# Creation of set
seett = {"chocolate","vanilla","pista"}
print(seett)
```

```python
#Duplicates not allowed
seett = {"chocolate","vanilla","pista","chocolate"}
print(seett)
```

```python
# The values True and 1 are considered the same value in sets
seett = {"chocolate","vanilla","pista",True,1,2}
print(seett)
```

```python
# A set with strings, integers and boolean values:
set1 = {"abc",34,True,78,"female",10.5}
print(set1)#note that in output it prints of any order as output cause set is unordered
```

```python
# To check membership
seett = {"chocolate","vanilla","pista"}

print("chocolate" in seett )
print("biscoff" in seett)
```

```python
# Once a set is created,you cannot replace its items,
# but you add new items.
seett = {"chocolate","vanilla","pista"}

seett.add("mango")
print(seett)
```

```python
#Remove Item
seett = {"chocolate","vanilla","pista"}

seett.remove("pista")

print(seett)
```

```python
# Set operations
A = {12,13,14,15,16}
B = {16,17,18,19,20}

print("Union:",A|B)
print("Intersection:",A&B)
print("Difference:",A-B)
print("Symmetric Difference:",A^B)
```

### Python frozenset


```python
# frozenset
x = frozenset({"chocolate","vanilla","pista"})
print(x)
print(type(x))
```

```python
#Adding or Removal of elements are not allowed in Frozenset
x = frozenset({"chocolate","vanilla","pista"})
x.remove("pista")
print(x)
```

```python
#Adding or Removal of elements are not allowed in Frozenset
x = frozenset({"chocolate","vanilla","pista"})
x.add("mango")
print(x)
```

Dictionary

they are used to store data values in key:value pairs.
they are ordered and unchangable also no duplication allowed

```python
# creation of dictionary
thisdict = {
    "brand":"Porche",
    "model":"GTR",
    "year": 1978
}
thisdict
```

```python
# Retriving values from the dictionary
thisdict = {
    "brand":"Porche",
    "model":"GTR",
    "year": 1978
}
print(thisdict["brand"])
```

```python
# Dictionary takes the latest value given for key. It will not allow duplication
thisdict = {
    "brand":"Porche",
    "model":"GTR",
    "year": 1978,
    "year": 2026
}
print(thisdict)
```

```python
# Length of the key-value pairs
len(thisdict)
```

```python
# Changing/Adding key:value in dictionary
car = {
    "brand":"Porche",
    "model":"GTR",
    "year": 1978
}
car["color"]="green"
x = car.values() #using this only values can be printed without the keys
print(x) #after the change
```

```python
#Retrival of values
student = {
    "roll": "169",
    "name":"Dev",
    "branch":"ENTC"
}
print("name:",student["name"])
```

```python
#Adding and Updating Elements
student["year"]="First"
student["name"]="Dev"

print(student)
```

```python
# Removing Elements
student.pop("branch")
print("After removal:",student)
```

'''4.A dictionary stores student name and marks.
Write a Python Program to find the student who scored the maximum marks.'''

```python
marks = {"Dev":98,"Asit":69,"Dristi":94}
topper = max(marks,key=marks.get)
print("Topper:",topper)
print("Marks:",marks[topper])
```
